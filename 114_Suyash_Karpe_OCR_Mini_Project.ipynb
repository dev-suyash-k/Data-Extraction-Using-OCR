```python
import os
import glob
import pytesseract
import re
import pandas as pd


class UserInput():
    '''This class take a input from user about which file information they want Aadhar Card or pancard.'''
    def welcome_note(self):
        print('Welcome to OCR - Here you convert your image into text.')
        print()
    def user_input(self):
        print('Press "1" for csv file of Aadhar Card information.')
        print('Press "2" for csv file of Pan Card information.')
        print('Press "3" for csv file of Both card information.')
        print()
        
        self.choice = int(input('Enter your choice here = '))
        return self.choice
    
## Class for extracting data from aadhar card and pancard images :-    
    
class Ocr():
    
    def Aadhar_card(self):
        '''This Method extract the data from aadhar card images.'''
        
        tesseract_path = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
        pytesseract.pytesseract.tesseract_cmd = tesseract_path
        path = r'C:\Users\shrir\Desktop\velocity\python practice\My_Assignment\adhar_card_imges\**'
        aadhar_file_list = glob.glob(path + '**.jpg')
        
        self.aadhar_file_name = []
        pattern = '[\w]{2,15}[.][jpg]{3}'
        for i in  aadhar_file_list:
            new_string = re.search(pattern,i)
            self.aadhar_file_name.append(new_string.group())

        self.adhar_number = []
        self.adhar_dob = []
       
        for i in  aadhar_file_list:
            text = pytesseract.image_to_string(i)
            num = re.findall('\d{3,4}\s\d{3,4}\s\d{3,4}',text)
            self.adhar_number.append(num)
            dob = re.findall('\d{1,2}[-/]\d{1,2}[-/]\d{4}',text)
            self.adhar_dob.append(dob)
        return self.adhar_number, self.adhar_dob
    
    def Pan_card(self):
        '''This Method extract the data from pan card images.'''
        
        tesseract_path = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
        pytesseract.pytesseract.tesseract_cmd = tesseract_path
        path = r'C:\Users\shrir\Desktop\velocity\python practice\My_Assignment\pan_card\**'
        pan_list = glob.glob(path + '**.jpg')
        
        self.new_pan_list = []
        pattern = '[\w]{2,15}[.][jpg]{3}'
        for i in pan_list:
            new_pan_string = re.search(pattern,i)
            self.new_pan_list.append(new_pan_string.group())
            
        self.pan_number = []
        self.pan_dob = []    
        
        for i in pan_list:
            pan_text = pytesseract.image_to_string(i)
            num = re.findall('[A-Z]{5}\d{4}[A-Z]{1}',pan_text)
            self.pan_number.append(num)
            dob = re.findall('\d{1,2}[-/]\d{1,2}[-/]\d{4}',pan_text)
            self.pan_dob.append(dob)
        return self.pan_number,self.pan_dob  

    def printing_dict(self):
        pan_card_dict = {'File Name' : self.new_pan_list, 
                         'Pan Number' : self.pan_number, 
                         'Date of Birth' : self.pan_dob } 
      

        aadhar_card_dict = {'File Name' :self.aadhar_file_name, 
                       'Aadhar Number' : self.adhar_number,
                       'Date of Birth' : self.adhar_dob } 
        
        return pan_card_dict,aadhar_card_dict
    
    def methods_calling(self):
        self.Aadhar_card()
        self.Pan_card()
        return self.printing_dict()
    
## Class for converting information to csv files :-    
    
class CsvFile():
    '''This class convert our information into CSV files.'''
    def __init__(self,adhar_dict,pan_dict):
        self.aadhar_info = adhar_dict
        self.pan_info = pan_dict
    def converting(self):
        df = pd.DataFrame(self.aadhar_info)
        df.to_csv('aadhar.csv',index = False)
        
        df = pd.DataFrame(self.pan_info)
        df.to_csv('pan.csv',index = False)
        
        return pd.read_csv("aadhar.csv"),pd.read_csv("pan.csv")
        
    
## Making object of class :-    
        
intro = UserInput()
intro.welcome_note()
user_choice = intro.user_input()
obj = Ocr()       
value = obj.methods_calling()
obj1 = CsvFile(value[0],value[1])
file = obj1.converting()

## printing of information according to user input :-


if user_choice == 1:
    print(f'Aadhar card information in the form of dictionary = \n\n{value[1]}')
    print()
    print(f'Content in Csv file of Aadhar card = \n\n{file[1]}' )
elif user_choice == 2:
    print(f'Pan card information in the form of dictionary = \n\n{value[0]}')
    print()
    print(f'Content in Csv file of Pan card = \n\n{file[0]}' )
else:
    print(f'Aadhar card information in the form of dictionary = \n\n{value[1]}')
    print()
    print(f'Content in Csv file of Aadhar card = \n\n{file[1]}' )
    print()
    print()
    print(f'Pan card information in the form of dictionary = \n\n{value[0]}')
    print()
    print(f'Content in Csv file of Pan card = \n\n{file[0]}' )
```
